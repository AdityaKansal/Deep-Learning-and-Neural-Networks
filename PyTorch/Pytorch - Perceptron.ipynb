```python
#importing libraries
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
import torch
from torch import nn
```

```python
#create dataset that is linearly separable
n_points = 100
centers = [[-1,-1],[1,1]]
X,Y = datasets.make_blobs(n_samples=n_points,random_state=123,centers=centers,cluster_std=0.6)
X_data = torch.from_numpy(X).float()
Y_data = torch.from_numpy(Y).float()
```

```python
#plotting datasets
plt.scatter(X[Y==0,0],X[Y==0,1])
plt.scatter(X[Y==1,0],X[Y==1,1])
plt.show()
```

```python
#### Using Class Object Approach
class Model(nn.Module):
    def __init__(self,input_size,output_size):
        super().__init__()
        self.model = nn.Linear(input_size,output_size)
        
    def forward(self,x):
        y = torch.sigmoid(self.model(x))
        return y
    def predict(self,x):
        prob = torch.sigmoid(self.model(x))
        if prob >= 0.5:
            return 1
        else: 
            return 0
```

```python
torch.manual_seed(123)
model = Model(2,1)
w,b=model.parameters()
print(w[0][0].item())
```

```python
criterion = nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(),lr=0.01)
```

```python
epochs = 1000
losses = []
for epoch in range(epochs):
    Y_pred = model.forward(X_data)
    loss = criterion(Y_pred,Y_data)
    losses.append(loss)
    print('epoch ',epoch,'loss is ',loss)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    
```
