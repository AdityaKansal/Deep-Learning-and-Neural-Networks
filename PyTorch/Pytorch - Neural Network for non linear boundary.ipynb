```python
#importing libraries
import torch
import numpy as np
from sklearn import datasets
from torch.nn import Linear 
import matplotlib.pyplot as plt
```

```python
#Creating datasets
n_points = 500
X,Y = datasets.make_circles(n_samples=n_points,random_state=123,noise=0.1,factor=0.4)
plt.scatter(X[Y==0,0],X[Y==0,1])
plt.scatter(X[Y==1,0],X[Y==1,1])
plt.plot()
X_data = torch.from_numpy(X).float()
Y_data = torch.from_numpy(Y).float()
```

```python
#creating Model architecture - one hidden layer with 4 neurons(4 linear separator)
class nn_model(torch.nn.Module):
    def __init__(self,input_size,h1_size,output_size):
        super().__init__()
        self.linear1 = Linear(input_size,h1_size)
        self.linear2 = Linear(h1_size,output_size)
    
    def forward(self,x):
        h1_out = torch.sigmoid(self.linear1(x))
        y_pred = torch.sigmoid(self.linear2(h1_out))
        return y_pred        
        
```

```python
#using model
model = nn_model(2,10,1)
def get_params():
    w1,b1,w2,b2 = model.parameters()
    print(w1,b1,w2,b1)
get_params()
```

```python
criterion = torch.nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(),lr = 0.01)
epochs = 500
losses = []
for epoch in range(epochs):
    Y_pred = model(X_data)
    loss = criterion(Y_pred,Y_data)
    print('epoch ',epoch,'loss ',loss)
    losses.append(loss)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python

def scatter_plot():
  plt.scatter(X[Y==0, 0], X[Y==0, 1])
  plt.scatter(X[Y==1, 0], X[Y==1, 1])

def plot_decision_boundary(X, y):
  x_span = np.linspace(min(X[:, 0]) -0.25, max(X[:, 0])+0.25)
  y_span = np.linspace(min(X[:, 1]) -0.25, max(X[:, 1])+0.25)
  xx, yy = np.meshgrid(x_span, y_span)
  grid = torch.Tensor(np.c_[xx.ravel(), yy.ravel()])
  pred_func = model.forward(grid)
  z = pred_func.view(xx.shape).detach().numpy()
  plt.contourf(xx, yy, z)
plot_decision_boundary(X, Y)
scatter_plot()
```
