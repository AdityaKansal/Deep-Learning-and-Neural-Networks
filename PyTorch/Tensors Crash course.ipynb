```python
#numpy to tensor and vice versa
x = np.array([1,2,3,4])
print(x.dtype)
y = torch.from_numpy(x)
print(y.dtype)
z = y.numpy()
print(z.dtype)
```

```python
#vectors operations
x = torch.tensor([1,2,3])
y = torch.tensor([1,2,3])
print(x+y)
```

```python
dot_product = torch.dot(x,y)
print(dot_product)
```

```python
x = torch.FloatTensor([1,2,3,4])
y = torch.exp(x)
print(x+1)
print(x*2)
print(y)
```

```python
#gradients
x = torch.tensor(1.0,requires_grad=True)
z = torch.tensor(2.0,requires_grad= False)
y = 3*x**2 + 4*x +5 +z**2 
```

```python
y.backward(retain_graph=True)
```
